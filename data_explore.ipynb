# MallPulse — Day 3: Istanbul Mall Data Exploration
Goal: understand the raw dataset shape before running `prep_data.py`.

````python
import pandas as pd

df = pd.read_csv('customer_shopping_data.csv')
print('Shape:', df.shape)
print('\nColumn types:')
print(df.dtypes)
print('\nFirst 5 rows:')
df.head()
````

````python
print('Null counts:')
print(df.isnull().sum())
````

````python
print('Unique malls:', df['shopping_mall'].unique())
print('\nUnique categories:', df['category'].unique())
````

````python
df['invoice_date'] = pd.to_datetime(df['invoice_date'], dayfirst=True)
print('Date range:', df['invoice_date'].min(), '→', df['invoice_date'].max())
print('\nRows per mall:')
print(df['shopping_mall'].value_counts())
````

````python
print('Rows per category:')
print(df['category'].value_counts())
print('\nPayment methods:')
print(df['payment_method'].value_counts())
print('\nGender split:')
print(df['gender'].value_counts())
````

````python
df['total'] = df['quantity'] * df['price']
print('Revenue by mall:')
print(df.groupby('shopping_mall')['total'].sum().sort_values(ascending=False).apply(lambda x: f'{x:,.0f}'))
````

````python
print('Revenue by category:')
print(df.groupby('category')['total'].sum().sort_values(ascending=False).apply(lambda x: f'{x:,.0f}'))
````

````python
print('Age distribution:')
print(df['age'].describe())
print('\nUnique customers:', df['customer_id'].nunique())
````
